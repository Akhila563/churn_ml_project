```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ',np.nan)
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].astype(object)
```

```python
df['TotalCharges'].fillna(0, inplace=True)
```

```python
df.info()
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
df.isnull().sum()
```

```python
df.loc[df['TotalCharges'].isnull() == True]
```

```python
df.dropna(how = 'any', inplace = True)
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].astype(object)
```

```python
df.info()
```

```python
df
```

```python
df.columns
```

```python
df = df.drop(columns=['customerID'])
```

```python
X = df.drop(columns=['Churn'])

y = df['Churn']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train.dtypes
```

```python
from sklearn.preprocessing import OneHotEncoder

X_train_cat = X_train.select_dtypes(include=['object'])
X_train_num = X_train.select_dtypes(include=['int64','float'])
X_test_cat = X_test.select_dtypes(include=['object'])
X_test_num = X_test.select_dtypes(include=['int64','float'])

encoder = OneHotEncoder()
x_train_cat_transformed = encoder.fit_transform(X_train_cat)
feature_names = encoder.get_feature_names_out(X_train_cat.columns)
X_train_cat_transformed_df = pd.DataFrame(x_train_cat_transformed.toarray(), columns=feature_names, index=X_train_cat.index)
X_train_cat_transformed_df
```

```python
## Test Data
x_test_cat_transformed = encoder.transform(X_test_cat)
feature_names = encoder.get_feature_names_out(X_test_cat.columns)
X_test_cat_transformed_df = pd.DataFrame(x_test_cat_transformed.toarray(), columns=feature_names, index=X_test_cat.index)
X_test_cat_transformed_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_num_transformed = scaler.fit_transform(X_train_num)
X_test_num_transformed = scaler.transform(X_test_num)
X_train_num_transformed_df = pd.DataFrame(scaler.fit_transform(X_train_num),
                                          columns = X_train_num.columns,
                                          index = X_train_num.index)
X_test_num_transformed_df = pd.DataFrame(scaler.transform(X_test_num),
                                         columns = X_test_num.columns,
                                         index = X_test_num.index)
X_test_num_transformed_df
```

```python
X_train_transformed = pd.concat([X_train_num_transformed_df,X_train_cat_transformed_df],axis=1)
X_test_transformed = pd.concat([X_test_num_transformed_df,X_test_cat_transformed_df],axis=1)

X_train_transformed
```

```python
X_test_transformed
```

```python
from sklearn import metrics
```

```python
%%time

from sklearn.linear_model import LogisticRegression  

regressor = LogisticRegression()
regressor.fit(X_train_transformed, y_train)

y_test_pred = regressor.predict(X_test_transformed)

#print('Accuracy of Logistic Regression',metrics.accuracy_score(y_test, y_test_pred))
#print('F1 score',metrics.f1_score(y_test, y_test_pred,pos_label='Yes'))
#print()
```

```python
%%time
print('Accuracy of Logistic Regression',metrics.accuracy_score(y_test, y_test_pred))
```

```python
print('F1 score',metrics.f1_score(y_test, y_test_pred,pos_label='Yes'))
print()
```
